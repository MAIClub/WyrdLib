```
import numpy as np
from google.colab.patches import cv2_imshow

class Image :
   
  
  def __init__ (self,image,stride,stride_x,stride_y,color_prefer):
    self.image=image
    self.stride_x=stride_x
    self.color_prefer=color_prefer
    self.stride_y=stride_y
    self.stride_x=stride_x
    stride_x=1
    stride_y=stride_x
    self.stride=stride
    stride=(stride_x,stride_y)
      
    
  def max_pooling (image,stride=(3,3),color_prefer=1):
    import cv2
    import matplotlib.pyplot as plt
    image= cv2.imread(image) 
    stride_x=stride[0]
    stride_y=stride[1]
    
    red = np.zeros((image.shape[0],image.shape[1]),dtype=int)
    green = np.zeros((image.shape[0],image.shape[1]),dtype=int)
    blue = np.zeros((image.shape[0],image.shape[1]),dtype=int)
                   
    white= np.zeros((image.shape[0],image.shape[1]),dtype=int)
    black= np.zeros((image.shape[0],image.shape[1]),dtype=int)
    
    colors=np.array((red,green,blue),dtype=int)
    wb=np.array((red,green,blue),dtype=int)
    
    for t in range(len(colors)):
      for i in range (image.shape[0]):
        for j in range (image.shape[1]):
          colors[t][i][j]=int(image[i][j][t])
          wb[t][i][j]=int((int(image[i][j][0])+int(image[i][j][1])+int(image[i][j][2]))//3)
    
    color_prefer_list=[wb,colors]
    color_mode=color_prefer_list[color_prefer]
    b=np.zeros((stride_x,stride_y),dtype=int)
    
    k=(image.shape[1]//b.shape[1])
    p=(image.shape[0]//b.shape[0])
    
    max_val_b=np.zeros((1,k),dtype=int)
    max_val_u=np.zeros((p,k),dtype=int)
    max_val_tot=np.zeros((3,p,k),dtype=int)
    u=np.zeros((k,stride_x,stride_y))
  
    reddy=np.zeros((p,k,stride_x,stride_y),dtype=int)
    greeny=np.zeros((p,k,stride_x,stride_y),dtype=int)
    bluey=np.zeros((p,k,stride_x,stride_y),dtype=int)
    colorsy=[reddy,greeny,bluey]
        
    x=0
    y=0

    total=np.zeros((3,p,k,stride_x,stride_y),dtype=int)
    
    for rgb_kinds in range (len(color_mode)):
      cur_col_mod=color_mode[rgb_kinds]
      l=colorsy[rgb_kinds]

      s=cur_col_mod.shape[0]
      d=cur_col_mod.shape[1]

      for gu in (0,stride_x):
        if (s%stride_x) != 0:
          cur_col_mod=np.delete(cur_col_mod,[s-1],axis=0)
          s-=1
        else:
          break
      for gr in (0,stride_y):
        if (d%stride_y) != 0:
          cur_col_mod=np.delete(cur_col_mod,[d-1],axis=1)
          d-=1
        else:
          break
    
      for ı in range (0,p):    
        for t in range (0,k):
          for i in range (0,b.shape[0]):
            for j in range (0,b.shape[1]):
              m=(x%stride_x)
              n=(y%stride_y)
              b[m][n]=cur_col_mod[x][y]
              y+=1        
            x+=1
            y-=(b.shape[1])
        
          max_val=np.amax(b)
          max_val_b[0][t]= max_val
          u[t]=b            
          x-=b.shape[0]
          y+=b.shape[1]
    
        max_val_u[ı]=max_val_b
        l[ı]=u        
        x+=b.shape[0]
        y-=((b.shape[1])*k)
    
      x=0
      y=0
    
      max_val_tot[rgb_kinds]=max_val_u
    
      total[rgb_kinds]=l
  
  
    max_red=max_val_tot[0]  
    max_green=max_val_tot[1]  
    max_blue=max_val_tot[2]
    rgb=np.dstack((max_red,max_green,max_blue))
    
    cv2_imshow(rgb)
```
